```python
# Plot a bar chart of the performance of ViT fine-tuning on 3 datasets

import matplotlib.pyplot as plt

# Data
datasets = ['Original(2.7k)', 'Cleaned(2.3k)', 'Enhanced(4k)']
accuracy = [0.9275, 0.9528, 0.9754]

# Plot bar chart
plt.bar(datasets, accuracy)
plt.xlabel('Datasets')
plt.ylabel('Accuracy')
plt.title('ViT fine-tuning on 3 datasets(4 epochs)')
plt.ylim(0.9, 1.0)

# Save plot
plt.savefig('vit_finetuning.png')

```
